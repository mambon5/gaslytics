```python
import requests
import json

# Replace with your actual values
username = "bkwenergie01"
password = "vyqpidi3Jv9lueS"
base_url = "https://apis.marketnews.com"  # Replace with the real API base URL

# Endpoint for obtaining the access token
url = f"{base_url}/api/auth/client/token"

# Request body
payload = {
    "username": username,
    "password": password
}

# Make the POST request
response = requests.post(url, json=payload)

# Check if request was successful
if response.status_code == 200:
    tokens = response.json()
    print("Access token:", tokens.get("access_token"))
    print("Expires in:", tokens.get("expires_in"), "seconds")
    print("Refresh token:", tokens.get("refresh_token"))
else:
    print("Error:", response.status_code, response.text)
```

```python
access_token = tokens.get("access_token")
```

```python
url = "https://apis.marketnews.com/api/v1/news/articles?section=us-gas-bullets"

headers = {
  "Content-Type": "application/json",
  "Authorization": f"Bearer {access_token}"
}

response = requests.get(url, headers=headers)

print(response.status_code)
print(json.dumps(response.json(), indent=2, ensure_ascii=False))
```
